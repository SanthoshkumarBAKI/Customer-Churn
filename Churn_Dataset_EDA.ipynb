**1.0.1 EDA Summary on Churn Dataset**
 **1. Customer Service Calls vs. Churn (Box Plot & Pie Chart)**
 • **Insight**: Customers who churn tend to contact customer service more frequently.

 • **Key Finding**: The box plot shows that churned customers usually have more customer service calls. The pie chart also reveals that churned customers are more likely to have made multiple calls, suggesting issues with the service.

 **2. Total Day Minutes by Histplot**

 • **Insight:** Customers with high daytime call minutes are more likely to churn.

 • **Key Finding:** The histplot shows that customers with excessive daytime call minutes may face higher charges, which could lead to dissatisfaction and increased churn.

 **3. Churn Rate by Voice Mail Plan (Count Plot)**

 • **Insight:** Customers with a voicemail plan are less likely to      churn.

 • **Key Finding:** The count plot shows that fewer churned customers have a voicemail plan,indicating that offering this feature can improve customer retention.

 **4. Total International Calls vs. Customer Service Calls (Box Plot)**

 • **Insight**: Customers who make more international calls also tend to contact customer service more often.

 • **Key Finding**: The box plot shows that customers with higher international call usage often have more customer service interactions, likely due to billing or plan issues.
